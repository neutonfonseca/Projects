```python
#loading yelp
yelpdata = pd.read_csv('yelpreviews_final2.csv', dtype={"phonenumber": str, "name": str, "userratingdate": str,
                                                       "userlocation": str, "delivery": str, "content": str,
                                                       "userlanguage": str, "overallrating": str, "price": str,
                                                       "userrating": str, "address": str, "typeoffood": str, "neighborhood": str})
```

```python
#Cleaning the database
yelpdata['neighborhood'] = yelpdata['neighborhood'].astype(str)
yelpdata['neighborhood'] = list(map(lambda s: s.strip(), yelpdata['neighborhood']))
yelpdata['name'] = list(map(lambda s: s.strip(), yelpdata['name']))
yelpdata['overallrating'] = list(map(lambda s: s[0:3], yelpdata['overallrating']))
yelpdata['overallrating'] = yelpdata['overallrating'].astype(float)
yelpdata['userrating'] = list(map(lambda s: s[0:3], yelpdata['userrating']))
yelpdata['userrating'] = yelpdata['userrating'].astype(float)
yelpdata['phonenumber'] = yelpdata['phonenumber'].astype(str)
yelpdata['phonenumber'] = list(map(lambda s: s.strip(), yelpdata['phonenumber']))
yelpdata['address'] = yelpdata['address'].astype(str)
yelpdata['address'] = list(map(lambda s: s.strip(), yelpdata['address']))
yelpdata = yelpdata.drop('typeoffood', 1)
yelpdata = yelpdata.drop('userlanguage', 1)
yelpdata['userlocation'] = yelpdata['userlocation'].astype(str)
yelpdata['userstate'] = yelpdata.userlocation.str.split(',').apply(lambda x: x[-1]).str.strip()
yelpdata['delivery'] = yelpdata['delivery'].astype(str)
yelpdata['delivery'] = list(map(lambda s: s.strip(), yelpdata['delivery']))
A = yelpdata.address.str.split(',').apply(lambda x: x[-2:-1])
yelpdata['restcity'] = ["".join(A[i]) for i in list(range(len(A)))]
```

```python
#Cleaning deeper:
yelpdata.loc[yelpdata['restcity'] == 'NY', 'restcity'] = 'New York'
yelpdata.loc[yelpdata['restcity'] == 'New York City', 'restcity'] = 'New York'
```

```python
yelpdata['borough'] = yelpdata['restcity']
yelpdata.loc[(yelpdata['borough'] == 'New York') |
             (yelpdata['borough'] == 'Roosevelt Island') |
             (yelpdata['borough'] == 'Harlem')
             , 'borough'] = 'Manhattan'
yelpdata.loc[yelpdata['borough'] == 'Tompkinsville', 'borough'] = 'Staten Island'
yelpdata.loc[(yelpdata['borough'] == 'Astoria') |
             (yelpdata['borough'] == 'Ridgewood') |
             (yelpdata['borough'] == 'Forest Hills') |
             (yelpdata['borough'] == 'Long Island City') |
             (yelpdata['borough'] == 'Flushing') |
             (yelpdata['borough'] == 'Corona') |
             (yelpdata['borough'] == 'Elmhurst') |
             (yelpdata['borough'] == 'Sunnyside') |
             (yelpdata['borough'] == 'Woodside') |
             (yelpdata['borough'] == 'Jackson Heights')
             , 'borough'] = 'Queens'
yelpdata.loc[(yelpdata['borough'] == '') & (yelpdata['neighborhood'] == 'Meatpacking District, Chelsea'), 'borough'] = 'Manhattan'
yelpdata.loc[(yelpdata['borough'] == '') & (yelpdata['neighborhood'] == 'Chelsea'), 'borough'] = 'Manhattan'
yelpdata.loc[(yelpdata['borough'] == '') & (yelpdata['neighborhood'] == 'Lower East Side'), 'borough'] = 'Manhattan'
yelpdata.loc[(yelpdata['borough'] == '') & (yelpdata['neighborhood'] == 'Upper West Side'), 'borough'] = 'Manhattan'
yelpdata.loc[(yelpdata['borough'] == '') & (yelpdata['neighborhood'] == 'Long Island City'), 'borough'] = 'Queens'
yelpdata = yelpdata.loc[(yelpdata['borough'] != 'Jersey City') & (yelpdata['borough'] != 'Edgewater')]
```

```python
yelpdata['year'] = yelpdata.userratingdate.str.split('/').apply(lambda x: x[-1]).str.strip() 
yelpdata['day'] = yelpdata.userratingdate.str.split('/').apply(lambda x: x[1]).str.strip() 
yelpdata['month'] = yelpdata.userratingdate.str.split('/').apply(lambda x: x[0]).str.strip() 
yelpdata.loc[yelpdata['month'] == "1", 'month'] = 'January'
yelpdata.loc[yelpdata['month'] == "2", 'month'] = 'February'
yelpdata.loc[yelpdata['month'] == "3", 'month'] = 'March'
yelpdata.loc[yelpdata['month'] == "4", 'month'] = 'April'
yelpdata.loc[yelpdata['month'] == "5", 'month'] = 'May'
yelpdata.loc[yelpdata['month'] == "6", 'month'] = 'June'
yelpdata.loc[yelpdata['month'] == "7", 'month'] = 'July'
yelpdata.loc[yelpdata['month'] == "8", 'month'] = 'August'
yelpdata.loc[yelpdata['month'] == "9", 'month'] = 'September'
yelpdata.loc[yelpdata['month'] == "10", 'month'] = 'October'
yelpdata.loc[yelpdata['month'] == "11", 'month'] = 'November'
yelpdata.loc[yelpdata['month'] == "12", 'month'] = 'December'
```

```python
yelpdata['short_address'] = yelpdata['address'].str.split(',').apply(lambda x: x[0]).str.strip()
```

```python
yelpdata['id'] = yelpdata['name'].astype(str) + ' - '+ yelpdata['short_address']
```

```python
#Creating the id to join the dataframe from yelp and seamless (using the 3 first letters of the restaurant and address)
yelprest = pd.DataFrame(yelpdata.groupby(['name','overallrating','address','delivery','price','borough','short_address']).userstate.count()).reset_index()
yelprest.columns = yelprest.columns.str.replace('userstate','number of reviews')
yelprest.name = yelprest.name.str.lower().str.strip()
yelprest['name_3'] = list(map(lambda s: s[:4], yelprest['name']))
yelprest['id'] = yelprest['name_3'].astype(str) + ' - '+ yelprest['short_address']
yelprest
```

```python
#Doing the same with the seamless dataframe
seamlessrest = pd.DataFrame(seamlessdata.groupby(['restaurant','overallrating','address','address_comp','deliverytime','deliveryontime','foodwasgood','orderaccurate','price','borough']).username.count()).reset_index()
seamlessrest.columns = seamlessrest.columns.str.replace('username','number of reviews')
seamlessrest.restaurant = seamlessrest.restaurant.str.lower().str.strip()
seamlessrest['restaurant_3'] = list(map(lambda s: s[:4], seamlessrest['restaurant']))
seamlessrest['id'] = seamlessrest['restaurant_3'].astype(str) + ' - '+ seamlessrest['address']
seamlessrest
```

```python
#Merging both dataframe
final = pd.merge(yelprest, seamlessrest, on = 'id')
final
final.columns = ['Y_Restaurant','Y_overall','Y_address','Y_delivery','Y_price','Y_borough','Y_shortaddress',
                'Y_reviews','Y_test1','id','S_Restaurant','S_overall','S_shortaddress','S_address','S_deliverytime',
                 'S_deliveryontime','S_foodwasgood','S_orderaccurate','S_price', 'S_borough','S_reviews','test2']
final.Y_Restaurant.count()
```

```python
#Comparting ratings from yelp and seamless
box = final[['Y_Restaurant','Y_borough','Y_overall','S_overall']]
box['S_overall'] = box['S_overall']/20
box.columns = ['Y_Restaurant','Y_borough','Yelp_rating','Seamless_rating']
box = pd.melt(box, id_vars = ['Y_Restaurant','Y_borough'], var_name='site', value_name='rating')
f, ax = plt.subplots(figsize=(18, 6))
sns.boxplot(x="Y_borough", y="rating", data=box, hue='site')
```

```python
#All types of ratings per borough
box2 = final[['Y_Restaurant','Y_borough','Y_overall','S_overall','S_deliveryontime','S_foodwasgood','S_orderaccurate']]
box2['S_overall'] = box2['S_overall']/20
box2['S_deliveryontime'] = box2['S_deliveryontime']/20
box2['S_foodwasgood'] = box2['S_foodwasgood']/20
box2['S_orderaccurate'] = box2['S_orderaccurate']/20
box2 = box2.loc[box2['Y_borough'] != 'Bronx']
box2.columns = ['Y_Restaurant','Y_borough','Yelp_overall','Seamless_overall','Seamless_deliveryontime','Seamless_foodwasgood','Seamless_orderaccurate']
box2 = pd.melt(box2, id_vars = ['Y_Restaurant','Y_borough'], var_name='site', value_name='rating')
f, ax = plt.subplots(figsize=(18, 6))
sns.boxplot(x="Y_borough", y="rating", data=box2, hue='site')
plt.legend(loc='lower right')
plt.xlabel('Borough')
```

```python
#Comparing the number of reviews from yelp and seamless
bar = final[['Y_Restaurant','Y_borough','Y_reviews','S_reviews']]
bar.columns = ['Y_Restaurant','Y_borough','Yelp_reviews','Seamless_reviews']
bar = bar.groupby('Y_borough', as_index = False).agg('sum').sort_values(by='Yelp_reviews', ascending=False)
bar = pd.melt(bar, id_vars = 'Y_borough', var_name='site', value_name='reviews')
f, ax = plt.subplots(figsize=(15, 6))
sns.barplot(x='Y_borough', y='reviews', data=bar, hue='site')
```

```python
#Is there significant difference of rating from yelp and seamless for restaurants with more than 1000 reviews on yelp
F1 = final[['Y_Restaurant','Y_overall','S_overall','Y_reviews','S_reviews']]
F1.columns = ['Y_Restaurant','Yelp_overall','Seamless_overall','Yelp_reviews','Seamless_reviews']
F1['Seamless_overall'] = F1['Seamless_overall']/20
F1 = F1.sort_values(by='Yelp_reviews', ascending=False)
F1 = F1.loc[F1['Yelp_reviews'] >= 1000]
F1
F1 = F1[['Y_Restaurant','Yelp_overall','Seamless_overall']]

F1 = pd.melt(F1, id_vars = 'Y_Restaurant', var_name='site', value_name='rating')
f, ax = plt.subplots(figsize=(15, 10))
sns.barplot(x="rating", y="Y_Restaurant", data=F1, hue='site')
ax.set(ylabel='Restaurants')
```

```python
#How is the difference in the number of reviews on each site
F2 = final[['Y_Restaurant','Y_overall','S_overall','Y_reviews','S_reviews']]
F2.columns = ['Y_Restaurant','Yelp_overall','Seamless_overall','Yelp_reviews','Seamless_reviews']
F2['Seamless_overall'] = F2['Seamless_overall']/20
F2 = F2.sort_values(by='Yelp_reviews', ascending=False)
F2 = F2.loc[F2['Yelp_reviews'] >= 1000]
F2
F2 = F2[['Y_Restaurant','Yelp_reviews','Seamless_reviews']]
F2 = pd.melt(F2, id_vars = 'Y_Restaurant', var_name='site', value_name='reviews')
f, ax = plt.subplots(figsize=(15, 10))
sns.barplot(x="reviews", y="Y_Restaurant", data=F2, hue='site')
ax.set(ylabel='Restaurants')
```
